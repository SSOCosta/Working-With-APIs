# Registered Domain Names Search

```python
import pandas as pd
import requests
import json
```

```python
# Registered Domain Names Search
domains_response = requests.get ('https://api.domainsdb.info/v1/domains/search?domain=Lego')
print("Domains:", domains_response.status_code)
```

```python
results = domains_response.json()
#results
```

```python
pd.set_option("display.max_rows", 0)
#Getting the list of dictionnaries into the dataframe
data = pd.DataFrame(results)
data.head(10)
```

```python
def flatten_column(data, column_name):
    flattened = pd.DataFrame(dict(data[column_name])).transpose()
    columns = [str(col) for col in flattened.columns]
    flattened.columns = [column_name + '_' + colname for colname in columns]
    data = pd.concat([data, flattened], axis=1)
    data = data.drop(column_name, axis=1)
    return data

data1 = flatten_column(data, 'domains')
data1
```
